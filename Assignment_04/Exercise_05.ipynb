```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x values
x = np.linspace(-5, 5, 1000001)

# Create an empty array for the function values
y = np.empty_like(x)

# Compute the function values based on the given conditions
negative_x = x < 0
y[negative_x] = x[negative_x] * np.sin(x[negative_x] ** 2)
y[~negative_x] = np.cos(x[~negative_x])
print(y[negative_x])
print([~negative_x])
# Plot the function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the Function f(x)')
plt.show()
```
