```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x values in the interval [0, 10]
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10

# Calculate the corresponding y values using the sine function
y = np.sin(x)

# Create the plot
plt.plot(x, y, label='sin(x)', color='red')

# Add labels and a legend
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x) for x in [0, 10]')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a grid of (x, y) values
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values using the function cos(x^2 + y^2)
Z = np.cos(X**2 + Y**2)

# Create a figure with two subplots: contour and surface
fig = plt.figure(figsize=(12, 6))

# Contour plot
ax1 = fig.add_subplot(121)
contour = ax1.contourf(X, Y, Z, levels=20, cmap='viridis')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Contour Plot of cos(x^2 + y^2)')
plt.colorbar(contour, ax=ax1)

# Surface plot
ax2 = fig.add_subplot(122, projection='3d')
surface = ax2.plot_surface(X, Y, Z, cmap='viridis')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('cos(x^2 + y^2)')
ax2.set_title('Surface Plot of cos(x^2 + y^2)')

# Add a color bar to the surface plot
fig.colorbar(surface, ax=ax2, shrink=0.5, aspect=10)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
