```python
from scipy.sparse import diags, eye, csc_matrix
from scipy.sparse.linalg import spsolve

def sparseGmats(p=0.54, N=10000):
    q = 1 - p
    P = diags([q*np.ones(N), p*np.ones(N)], offsets=[-1, 1], shape=(N+1, N+1), format='dia')
    A = [0, N]
    B = list(range(1, N))
    P = csc_matrix(P)
    I_PBB = (eye(N-1, format='csc') - P[B][:, B]).tocsc()
    PBA = P[B][:, A].tocsc()
    return I_PBB, PBA

def ruinG(p=0.54, N=10000):
    I_PBB, PBA = sparseGmats(p, N)
    return spsolve(I_PBB, PBA[:, 0])

# Example usage
winning_probability = 0.54
total_games = 10000
probabilities_of_ruin = ruinG(winning_probability, total_games)
print(probabilities_of_ruin)
```

```python
# Call the ruinG function with N=10
N = 10
probabilities_of_ruin = ruinG(N=N)

# Print the probabilities of ruin for each starting state
print("Probabilities of G's ruin for each starting state:")
print(probabilities_of_ruin)
```

```python
import matplotlib.pyplot as plt

def find_prob(hs):
    j = 0
    for i in hs:
        if i - 0.1 <= 1e-15:
            break
        j += 1
    return j

def plot_prob(n):
    fig = plt.figure()
    ax = plt.gca()

    for i in range(1, n + 1):
        hs = ruinG(N=i * 10)
        pr = find_prob(hs)
        ax.scatter(pr, i * 10, label=f'N={i * 10}')

    ax.set_xlabel("Probability of G's ruin")
    ax.set_ylabel('Starting state index')
    ax.legend()
    plt.show()

```

```python
plot_prob(20)
```

```python
def probability_of_ruin(R, p, q):
    if R <= 0:
        return 1
    if R >= 1000:  # Assuming the company won't stop if the reserve becomes very high
        return 0
    return q * probability_of_ruin(R - 200, p, q) + p * probability_of_ruin(R + 100, p, q)

initial_reserve = 1000
p = 1 - 0.46
q = 0.46

probability_ruin = probability_of_ruin(initial_reserve, p, q)
print(f"Probability of ruin: {probability_ruin:.6f}")
```

```python
def find_required_reserve(p, q, target_probability):
    R = 1000
    while True:
        probability_ruin = probability_of_ruin(R, p, q)
        if probability_ruin < target_probability:
            return R
        R += 100  # Increment the reserve by $100

target_probability = 0.001  # 0.1%
required_reserve = find_required_reserve(p, q, target_probability)
print(f"Required reserve for P(ruin) < 0.1%: ${required_reserve}")
```
